```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("global_cybersecurity.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={
        'Country': 'country',
        'Year': 'year',
        'Attack Type': 'attack_type',
        'Target Industry': 'industry',
        'Financial Loss (in Million $)': 'financial_loss_millions',
        'Number of Affected Users': 'affected_users',
        'Attack Source': 'attack_source',
        'Security Vulnerability Type': 'vulnerability_type',
        'Defense Mechanism Used': 'defense_mechanism',
        'Incident Resolution Time (in Hours)': 'resolution_hours'
    })
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['year'] = pd.to_datetime(df['year'], format='%Y')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['financial_loss_millions'] = pd.to_numeric(df['financial_loss_millions'], errors='coerce')
df['affected_users'] = pd.to_numeric(df['affected_users'], errors='coerce')
df['resolution_hours'] = pd.to_numeric(df['resolution_hours'], errors='coerce')
```

```python
df.to_csv("cyberthreat_clean.csv")
```
